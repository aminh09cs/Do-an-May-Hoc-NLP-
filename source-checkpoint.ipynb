# 4. Chuẩn bị dữ liệu cho các thuật toán Machine Learning.

##### 4.1. Xử lý ngoại lệ (Identifying Outliers with Interquartile Range (IQR))

```python
Q1 = df.quantile(0.25) #Phần tử ở vị trí 25% dữ liệu mỗi cột
Q3 = df.quantile(0.75) #Phần tử ở vị trí 75% dữ liệu mỗi cột
IQR = Q3 - Q1
IQR
```

```python
#Ngoại lệ dưới(Nếu mỗi giá trị trong cột < giá trị này thì nó là 1 ngoại lệ)
low = df.quantile(0.25) - (IQR * 3) # Công thức có thể thay 3 thành 1.5, 2, 4,..
low
```

```python
#Ngoại lệ trên(Nếu mỗi giá trị trong cột > giá trị này thì nó là 1 ngoại lệ)
up = df.quantile(0.75) + (IQR * 3)
up
```

```python
#Đếm số giá trị ngoại lệ
num = [var for var in df.columns if df[var].dtype!='O']
for col in num:
    IQR = df[col].quantile(0.75) - df[col].quantile(0.25)
    low = df[col].quantile(0.25) -  IQR * 3
    up  = df[col].quantile(0.75) +  IQR * 3 
    
    min = df[col][df[col] < low].count()
    max = df[col][df[col] > up].count()
    print("Côt {} có số lượng các giá tị ngoại lệ là {} giá trị".format(col, min + max))
```

```python
#Chỉ xử lý những cột có giá trị ngoại lệ tương đối lớn
#Rainfall, Evaporation, WinGustSpeed,WinSpeed9am, WinSpeed3pm
df["Rainfall"] = np.where(df["Rainfall"] > 3.2, 3.2, df["Rainfall"])
df["Rainfall"] = np.where(df["Rainfall"] < -2.4 , -2.4, df["Rainfall"])

df["Evaporation"] = np.where(df["Evaporation"] > 21.8, 21.8, df["Evaporation"])
df["Evaporation"] = np.where(df["Evaporation"] < -11.8 , -11.8, df["Evaporation"])

df["WindGustSpeed"] = np.where(df["WindGustSpeed"] > 99, 99, df["WindGustSpeed"])
df["WindGustSpeed"] = np.where(df["WindGustSpeed"] < -20 , -20, df["WindGustSpeed"])

df["WindSpeed9am"] = np.where(df["WindSpeed9am"] > 55, 55, df["WindSpeed9am"])
df["WindSpeed9am"] = np.where(df["WindSpeed9am"] < -29 , -29, df["WindSpeed9am"])

df["WindSpeed3pm"] = np.where(df["WindSpeed3pm"] > 57, 57, df["WindSpeed3pm"])
df["WindSpeed3pm"] = np.where(df["WindSpeed3pm"] < -20 , -20, df["WindSpeed3pm"])
df.describe()
```

##### 4.2. Xử lý dữ liệu số bị thiếu

```python
#Thay thế những giá trị NaN bằng giá trị trung bình của cột đó
num = [var for var in df.columns if df[var].dtype!='O']
for col in num:
    col_median=df[col].median()
    col_median = round(col_median, 2)
    print("Cột {} có giá trị trung bình là: {}".format(col, col_median))
    df[col].fillna(col_median, inplace=True)
```

##### 4.3. Xử lý dữ liệu phân loại bị thiếu

```python
#Thay thế những giá trị còn thiếu trong cột bằng giá trị xuất hiện nhiều nhất
cate = [var for var in df.columns if df[var].dtype=='O']
for col in cate:
    print("Cột {} có giá trị xuất hiện nhiều nhất là: {}".format(col, df[col].mode()[0]))
    df[col].fillna(df[col].mode()[0], inplace=True)
```

##### 5.1. Chia hình thành 2 tập train và test

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns ="RainTomorrow")
y = df["RainTomorrow"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

##### 5.2 MÔ HÌNH LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression(solver='liblinear', random_state=0, max_iter=1000)
logit.fit(X_train, y_train)
```

```python
#Xem kết quả dự đoán của mô hình
y_pred = logit.predict(X_test)
y_pred #0 là không mưa, 1 là có mưa
```

```python
from sklearn.metrics import accuracy_score #Độ chính xác với tham số mặc định C=1
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Tỷ lệ dự đoán chính xác của mô hình test là: {}".format(accuracy))
```

```python
#So sánh độ chính xác của tập train và tập test
y_pred_train = logit.predict(X_train)
print("Tỷ lệ dự đoán chính xác của mô hình train là: {}".format(accuracy_score(y_train, y_pred_train)))
```

##### 5.2.1. Thay đổi tham số C

```python
logit_C = LogisticRegression(solver='liblinear', random_state=0, C=1e9)
logit_C.fit(X_train, y_train)

for i in np.arange(0,10, 1):
    logit_C = LogisticRegression(solver='liblinear', random_state=0, C=np.power(10, i))
    logit_C.fit(X_train, y_train)
    accuracy=accuracy_score(y_test, logit_C.predict(X_test))
    print("Độ đo với C = {} có kết quả là: {}".format(np.power(10, i), accuracy))
```

```python
lst_C = [0.00001, 0.0001, 0.001, 0.01, 0.1]
for i in lst_C:
    logit_C = LogisticRegression(solver='liblinear', random_state=0, C=i)
    logit_C.fit(X_train, y_train)
    accuracy=accuracy_score(y_test, logit_C.predict(X_test))
    print("Độ đo với C = {} có kết quả là: {}".format(np.power(10, i), accuracy))
```

##### Nhận xét:
